```python
import numpy as np
import matplotlib.pyplot as plt

index_1 = np.array([0.001204, 0.0028, 0.005992, 0.012376, 0.025116, 0.050624, 0.10164])
index_2 = np.array([0.000136, 0.000315, 0.000675, 0.001396, 0.002835, 0.005712, 0.011469])

x_vals = np.arange(len(index_1))

plt.figure(figsize=(8, 4))
plt.plot(x_vals, index_1, 'o-', label='index_1')
plt.yscale('log')
plt.xlabel('Index')
plt.ylabel('Value (log scale)')
plt.title('Plot of index_1 with Log Y Axis')
plt.grid(True, which="both", ls="--")
plt.legend()
plt.tight_layout()
plt.show()


plt.figure(figsize=(8, 4))
plt.plot(x_vals, index_2, 's-', label='index_2', color='orange')
plt.yscale('log')
plt.xlabel('Index')
plt.ylabel('Value (log scale)')
plt.title('Plot of index_2 with Log Y Axis')
plt.grid(True, which="both", ls="--")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np


original_index_1 = np.array([0.001204, 0.0028, 0.005992, 0.012376, 0.025116, 0.050624, 0.10164])
ASAP_index_1 = np.array([0.005, 0.010, 0.020, 0.040, 0.080, 0.160, 0.320 ])

log_ratios = np.log10(original_index_1 / original_index_1[0])
log_ratios -= log_ratios[0]  


new_start = 0.00166 * 0.9   # = 0.001494
new_end = 0.0715 * 1.1      # = 0.07865


log_start = np.log10(new_start)
log_end = np.log10(new_end)


new_logs = log_start + (log_end - log_start) * (log_ratios / log_ratios[-1])
new_index = 10 ** new_logs


print("✅ ：")
for i, val in enumerate(new_index):
    print(f"index[{i}] = {val:.6f}")

# 画图
x = np.arange(len(original_index_1))

plt.figure(figsize=(8, 5))
plt.plot(x, original_index_1, 'ro-', label='Original index_1')
plt.plot(x, new_index, 'bo--', label='Simulated index')
plt.plot(x, ASAP_index_1, 'go--', label='ASAP7')
plt.yscale('log')
plt.xlabel('Index Position')
plt.ylabel('Delay (log scale)')
plt.title('Original vs New index with Same Shape')
plt.grid(True, which="both", linestyle='--')
plt.legend()
plt.tight_layout()
plt.show()
```
